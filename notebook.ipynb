**Zero-shot vs few-shot:** pick one task (e.g. classify sentiment, or extract fields from a sentence). Run it
zero-shot, then with 2-3 examples. Compare the results

**Zero-shot**

````python
from summarizer import chat
````

````python
#example is from the amazon review database
example="Good Quality Dog Food,I have bought several of the Vitality canned dog food products and have found them all to be of good quality. The product looks more like a stew than a processed meat and it smells better. My Labrador is finicky and she appreciates this product better than  most."

zero_shot_message = [
        {"role": "system", "content": "Classify the sentiment of the review as Positive or Negative."},
        {"role": "user", "content": example},
    ]

print(chat(message=zero_shot_message, temperature=0))
#temperature is 0 so that the model is rigid
````

**Few-shot**

````python
few_shot_message = [
        {"role": "system", "content": "Classify the sentiment of the sentence as Positive, or Negative."},
        {"role": "user", "content": "Delight says it all,This is a confection that has been around a few centuries.  It is a light, pillowy citrus gelatin with nuts - in this case Filberts. And it is cut into tiny squares and then liberally coated with powdered sugar.  And it is a tiny mouthful of heaven.  Not too chewy, and very flavorful.  I highly recommend this yummy treat.  If you are familiar with the story of C.S. Lewis' The Lion, The Witch, and The Wardrobe - this is the treat that seduces Edmund into selling out his Brother and Sisters to the Witch."},
        {"role": "assistant", "content": "Positive"},
        {"role": "user", "content": "I purchased the Mango flavor, and to me it doesn't take like Mango at all.  There is no hint of sweetness, and unfortunately there is a hint or aftertaste almost like licorice.  I've been consuming various sports nutrition products for decades, so I'm familiar and have come to like the taste of the most of the products I've tried.  The mango flavor is one of the least appealing I've tasted.  It's not terrible, but it's bad enough that I notice the bad taste every sip I take."},
        {"role": "assistant", "content": "Negative"},
        {"role": "user", "content": example},
    ]

print(chat(message=few_shot_message, temperature=0))
#temperature is 0 so that the model is rigid
````

***Zero-shot vs few-shot comparision:***
* Zero-shot: give the instruction with no examples. Works well for simple tasks.
* Few-shot: give 2 examples before your query. Improves accuracy and format consistency, especially on smaller models.

The model used is relatively small, but the task itself is simple which makes the performance similar for both methods.

**Parameters:** run the SAME prompt at temperature 0, 0.7, and ~1.2 (and try one top_p change). Show the
outputs.

Temperature controls creativity for a model, so the prompt that I will test it with will need to have a room for creativity.

````python
prompt = [
        {"role": "system", "content": "You are a zoologist in a fictional setting with fantastical animals."},
        {"role": "user", "content": "Write a short paragraph describing your favorite animal in a short sentence."},
    ]
````

````python
#temperature = 0

print("Temperature=0:\n"+chat(prompt, temperature=0, max_tokens=100))
````

````python
#temperature = 0.7

print("Temperature=0.7:\n"+chat(prompt, temperature=0.7, max_tokens=100))
````

````python
#temperature = 1.2

print("Temperature=1.2:\n"+chat(prompt, temperature=1.2, max_tokens=100))
````

````python
#top_p=0.85

print("Top_p=0.85:\n"+chat(prompt, top_p=0.85, max_tokens=100))
````

Using different temperatures 0, 0.7, and 1.2 to let us compare the different outputs and the level of creativity shown at each temperature, and a run with the default temperature and a top_p at 0.85. At temperature 0 the output stayed unchanged across different runs, temperature 0.7 and 1.2 each show more variety and randomness across each run, making the output more creative but it also has a chance of being less coherent. The top_p 0.85 is lower than the default top_p closer to the low-temperature runs, this is because it shrinks the sampling pool.

**Structured output:** get the model to reliably return valid JSON for a small task (e.g. {"name": ..., "age": ...})

````python
import json
````

````python
person1 = "John Doe is a 30-year-old software developer."
prompt_message = f'{{"prompt": "Extract the personal information of the given person in valid JSON format, no markdown, no code fences, no ```json, no explanation, no extra text before or after:\n\n- Name (string)\n- Age (integer)\n- Occupation (string, optional)\n\n**Person:** {person1}"}}'

personal_info_message=[
    {"role": "system", "content": "You are a detailed information provider."},
    {"role": "user", "content": prompt_message}
]

resp=chat(message=personal_info_message, temperature=0)

print(resp)
````

````python
json.loads(resp)
````

Structured output: this section should extract personal information and return them as a valid JSON, the information includes name, age, and occupation. At first the model was outputting the JSON format with markdowns, this caused an issue when converting the str output into JSON, so I added a strict system prompt to ensure clean convertion. The temperature is set at 0 to keep the output consistent.
